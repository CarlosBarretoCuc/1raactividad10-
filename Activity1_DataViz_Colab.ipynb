## 4. Visualization 1 – Retention Rate Trend Over Time

```python
retention_by_year = df.groupby('Year')['Retention Rate (%)'].mean().reset_index()

fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(retention_by_year['Year'], retention_by_year['Retention Rate (%)'],
        marker='o', linewidth=2.5, color='#2196F3', label='Retention Rate')
ax.fill_between(retention_by_year['Year'], retention_by_year['Retention Rate (%)'],
                alpha=0.15, color='#2196F3')
ax.set_title('Retention Rate Trend Over Time', fontsize=14, fontweight='bold')
ax.set_xlabel('Year'); ax.set_ylabel('Retention Rate (%)')
ax.set_ylim(80, 95)
for _, row in retention_by_year.iterrows():
    ax.annotate(f"{row['Retention Rate (%)']:.0f}%",
                xy=(row['Year'], row['Retention Rate (%)']),
                xytext=(0, 8), textcoords='offset points', ha='center', fontsize=9)
plt.tight_layout()
plt.savefig('retention_trend.png', dpi=120)
plt.show()
```

## 5. Visualization 2 – Student Satisfaction Scores by Year

```python
sat_by_year = df.groupby('Year')['Student Satisfaction (%)'].mean().reset_index()

fig, ax = plt.subplots(figsize=(10, 4))
sns.lineplot(data=sat_by_year, x='Year', y='Student Satisfaction (%)',
             marker='s', linewidth=2.5, color='#4CAF50', ax=ax)
ax.fill_between(sat_by_year['Year'], sat_by_year['Student Satisfaction (%)'],
                alpha=0.12, color='#4CAF50')
ax.set_title('Student Satisfaction Scores by Year', fontsize=14, fontweight='bold')
ax.set_xlabel('Year'); ax.set_ylabel('Satisfaction (%)')
ax.set_ylim(70, 95)
plt.tight_layout()
plt.savefig('satisfaction_trend.png', dpi=120)
plt.show()
```

## 6. Visualization 3 – Spring vs Fall Comparison

```python
term_avg = df.groupby('Term')[['Enrolled', 'Retention Rate (%)', 'Student Satisfaction (%)']].mean()
print(term_avg)

fig, axes = plt.subplots(1, 3, figsize=(14, 4))
metrics = ['Enrolled', 'Retention Rate (%)', 'Student Satisfaction (%)']
colors  = ['#FF9800', '#2196F3', '#4CAF50']
for ax, metric, color in zip(axes, metrics, colors):
    term_avg[metric].plot(kind='bar', ax=ax, color=[color, color+'99'],
                          edgecolor='white', width=0.5)
    ax.set_title(metric, fontweight='bold')
    ax.set_xlabel('')
    ax.set_xticklabels(term_avg.index, rotation=0)
    for p in ax.patches:
        ax.annotate(f'{p.get_height():.1f}',
                    (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='bottom', fontsize=10)
fig.suptitle('Spring vs Fall – Key Metrics Comparison', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('spring_vs_fall.png', dpi=120)
plt.show()
```

## 7. Visualization 4 – Department Enrollment Distribution (Donut)

```python
dept_totals = {
    'Engineering': df['Engineering Enrolled'].sum(),
    'Business':    df['Business Enrolled'].sum(),
    'Arts':        df['Arts Enrolled'].sum(),
    'Science':     df['Science Enrolled'].sum(),
}
fig, ax = plt.subplots(figsize=(6, 6))
wedges, texts, autotexts = ax.pie(
    dept_totals.values(), labels=dept_totals.keys(),
    autopct='%1.1f%%', startangle=140,
    colors=['#2196F3', '#FF9800', '#4CAF50', '#9C27B0'],
    wedgeprops=dict(width=0.55))
ax.set_title('Total Enrollment by Department (2015–2024)', fontweight='bold')
plt.tight_layout()
plt.savefig('dept_enrollment.png', dpi=120)
plt.show()
```

## 8. Admissions Funnel – Applications vs Admitted vs Enrolled

```python
funnel = df.groupby('Year')[['Applications', 'Admitted', 'Enrolled']].mean().reset_index()

fig, ax = plt.subplots(figsize=(11, 4))
ax.plot(funnel['Year'], funnel['Applications'], marker='o', label='Applications', color='#E91E63')
ax.plot(funnel['Year'], funnel['Admitted'],     marker='s', label='Admitted',     color='#FF9800')
ax.plot(funnel['Year'], funnel['Enrolled'],     marker='^', label='Enrolled',     color='#2196F3')
ax.set_title('Admissions Funnel Over Time', fontsize=14, fontweight='bold')
ax.set_xlabel('Year'); ax.set_ylabel('Students')
ax.legend()
plt.tight_layout()
plt.savefig('admissions_funnel.png', dpi=120)
plt.show()
```

## 9. Key Findings Summary

| Finding | Detail |

| Retention improving | From 85% (2015) to 90% (2024) — a 5 pp gain |
| Satisfaction improving | From 78% (2015) to 88% (2024) |

| Engineering dominates | ~39% of all enrolled students across the period |
| Spring = Fall | Both terms show virtually identical averages |
